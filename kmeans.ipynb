# Библиотеки


```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

# Данные


```python
data2 = pd.read_excel("data1.xls", skiprows=3)
data2.head(5)
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(data2):
    # Drop column: 'Unnamed: 2'
    data2 = data2.drop(columns=['Unnamed: 2'])
    # Drop column: 'Unnamed: 0'
    data2 = data2.drop(columns=['Unnamed: 0'])
    # Drop rows with missing data across all columns
    data2 = data2.dropna()
    return data2

data2_clean = clean_data(data2.copy())
data2_clean.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(data2_clean):
    # Rename column 'Unnamed: 1' to 'Место'
    data2_clean = data2_clean.rename(columns={'Unnamed: 1': 'Место'})
    return data2_clean

data2_clean_1 = clean_data(data2_clean.copy())
data2_clean_1.head()
```

```python
data2_clean_1 = data2_clean_1.drop(0, axis=0)
data2_clean_1.head(5)
```

```python
X = data2_clean_1.drop(['Место'], axis=1)
Y = data2_clean_1['Место']
X.head(5)
```

```python
Y.head(5)
```

# **_Создание модели_**


```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
len(Y)
```

```python
print(Y_train.shape[0], X_train.shape[0])
```

```python
kmeans = KMeans(n_clusters=Y_train.shape[0], random_state=42, n_init="auto").fit(X_train)
kmeans.labels_
```

```python
kmeans.predict(X_test.iloc(axis=0)[[33]])
```

```python
kmeans2 = KMeans(n_clusters=25, random_state=0, n_init="auto").fit(X)
kmeans2.labels_
```

```python
kmeans2.labels_[0]
```

```python
Y.iloc(axis=0)[0]
```

```python
kmeans2.predict(X.iloc(axis=0)[[34]])
```

```python
kmeans2
```

```python
semi_ans = kmeans2.labels_
```
